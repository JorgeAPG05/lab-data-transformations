```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
# The line down below is needed to prevent matplotlib to open the graph in a seaparate popup window. 
%matplotlib inline
```

```python
data = pd.read_csv('marketing_customer_analysis.csv')
display(data)
print(data.dtypes)
```

```python
numerical_columns = data.select_dtypes(include=[np.number])
display(numerical_columns)
```

```python
categorical_columns = data.select_dtypes(include=[object])
display(categorical_columns)
```

```python
#Now we will try to check the normality of the numerical variables visually
#Use seaborn library to construct distribution plots for the numerical variables
#Use Matplotlib to construct histograms
#Do the distributions for different numerical columns look symmetrical? Compute the skewness for each, 
#and add a comment with your findings.
```

```python
for col in numerical_columns.columns:
    sns.displot(numerical_columns[col], kde=True)
    plt.title(f"Distribution of {col}")
    plt.show()
```

```python
numericalskewness= numerical_columns.skew()
print(numericalskewness)
```

```python
''' 
Customer Lifetime Value is extremely skewed to the right, which indicates that this column has a lot of positive high values

Income is also skewed to the right, although less so.

Monthly premium auto is also extremely skewed to the right

Months since Last claim is moderately skewed to the right (so there are many customers that haven't made a claim in a relati-
vely long period of time

Months since Policy Inception has a pretty low skew, meaning that it follows a normal distribution centered around 0

Number of open complaints is also extremely skewed to the right, same although is more moderate, with Number of Policies and
the Total Claim amount

```
